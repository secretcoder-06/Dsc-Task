```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("2021 New Coder Survey.csv",low_memory=False)
```

```python
df.head()
```

List all column names

Check the structure and summary statistics

```python
df.info()
df.describe()
```

Check data types and count missing values for each column

```python
df.isnull().sum()
```

Plot histogram for a numerical feature


```python
plt.figure(figsize=(12, 6))
df['22. About how much money did you earn last year from any job or employment (in US Dollars)? '].value_counts().plot(kind='bar')
plt.xlabel('Annual Salary (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Annual Salary')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Analyze categorical features with count plots

```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.bar(['No', 'Yes'], df['37. Do you financially support any dependents, older relatives or relatives with disabilities?'].value_counts())
plt.xlabel('Financially Supports Dependents')
plt.ylabel('Count')
plt.title('Distribution of Financial Support for Dependents')
plt.show()
```

the histograms for 'Hours Spent Learning per Week.

```python
plt.figure(figsize=(10, 6))
df['7. About how many hours do you spend learning each week?'].hist(bins=20)
plt.xlabel('Hours per Week')
plt.ylabel('Frequency')
plt.title('Distribution of Hours Spent Learning per Week')
plt.show()
```

the histograms for 'Hours Spent Learning as per Age.

```python
plt.figure(figsize=(10, 6))
df['23. How old are you?'].hist(bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```
